Get List of Nasdaq Listed Companies

```python
import pandas as pd
from yahoo_fin import stock_info as si
import yfinance as yf
import finviz

symbols_df = pd.DataFrame( si.tickers_nasdaq() )
symbols_df.head()
```

```python
print(f"Total Listed Companies: {len(symbols_df)}")
```

Get Data of Nasdaq Companies & store into csv file

```python
for idx,symbol in enumerate(symbols_df[0].tolist()):
    ticker = yf.Ticker(symbol)
    df = ticker.history(start="2011-01-01", end="2022-12-31", interval="1d")
    df['Timestamp'] = pd.to_datetime(df.index)
    df = df[['Timestamp','Open','High','Low','Close','Volume']]
    df.to_csv(f'Dataset/{symbol} 2011_2022.csv', index=False)
    print(f'Total Download Tikers {idx+1}/{len(symbols_df)}')

    
```

Get Fundamental Data of campanies

```python
df_list = []
for idx,symbol in enumerate(symbols_df[0].tolist()):
    try:
      df_list.append(finviz.get_stock(symbol))
      print(f'Total Download Tikers Fundamental Data {idx+1}/{len(symbols_df)}')
    except:
      print(f'error {symbol}')

df= pd.DataFrame(df_list)
df.to_csv('Fundamental Data of Nasdaq Companies.csv')
```
